```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from sklearn.metrics.pairwise import cosine_similarity
import networkx as nx
import matplotlib.pyplot as plt

# Example User-Item Interaction Data (users x items)
data = {
    'Milk': [1, 0, 1, 1, 0],
    'Bread': [1, 1, 1, 1, 1],
    'Butter': [0, 1, 1, 1, 0],
    'Cheese': [1, 0, 0, 1, 1],
    'Beer': [0, 1, 1, 0, 0]
}

user_data = {
    'user1': [1, 1, 0, 1, 0],
    'user2': [1, 0, 1, 0, 0],
    'user3': [0, 1, 1, 0, 1],
    'user4': [1, 1, 1, 0, 0],
    'user5': [0, 0, 1, 1, 1]
}

df = pd.DataFrame(data)  # Items interaction matrix
user_df = pd.DataFrame(user_data)  # User interaction matrix (user-item)

# Example Item Metadata (for content-based filtering)
item_metadata = {
    'Milk': {'category': 'Dairy', 'brand': 'BrandA'},
    'Bread': {'category': 'Bakery', 'brand': 'BrandB'},
    'Butter': {'category': 'Dairy', 'brand': 'BrandA'},
    'Cheese': {'category': 'Dairy', 'brand': 'BrandC'},
    'Beer': {'category': 'Beverage', 'brand': 'BrandD'}
}
metadata_df = pd.DataFrame.from_dict(item_metadata, orient='index')

# Apply Apriori algorithm to find frequent itemsets and generate association rules
def apply_apriori_algorithm(df, support=0.5):
    frequent_itemsets = apriori(df, min_support=support, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
    return rules

# Function to recommend products for new items based on association rules
def recommend_for_new_product(new_item_id, df, support=0.5):
    # Apply Apriori algorithm to generate association rules
    rules = apply_apriori_algorithm(df, support)

    # Find rules where the new product is in the antecedent or consequent
    recommended_products = set()

    for _, rule in rules.iterrows():
        # Check if the new item is in the antecedent or consequent
        if new_item_id in rule['antecedents']:
            recommended_products.update(rule['consequents'])
        elif new_item_id in rule['consequents']:
            recommended_products.update(rule['antecedents'])

    # Convert the set of recommended products into a list and return it
    return list(recommended_products)

# Example: Recommend products for a new product 'Butter' using the association rules
new_product = 'Butter'  # New product that needs recommendations
recommended_items = recommend_for_new_product(new_product, df, support=0.5)

print(f"Recommended products for the new item '{new_product}':")
print(recommended_items)

# Optional: Visualize the association rules network
def plot_association_rules(rules):
    """Visualize association rules using NetworkX"""
    G = nx.Graph()

    for _, rule in rules.iterrows():
        antecedent = tuple(rule['antecedents'])
        consequent = tuple(rule['consequents'])
        G.add_edge(antecedent, consequent, weight=rule['lift'])

    pos = nx.spring_layout(G, seed=42)
    nx.draw_networkx_nodes(G, pos, node_size=500, node_color='skyblue')
    nx.draw_networkx_edges(G, pos, width=2.0, alpha=0.7)
    nx.draw_networkx_labels(G, pos, font_size=10)

    plt.title('Association Rules Network')
    plt.show()

# Visualizing the association rules (optional)
rules = apply_apriori_algorithm(df, support=0.5)
plot_association_rules(rules)
```
